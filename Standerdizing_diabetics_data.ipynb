##**Imports**##

```python
import pandas as pd
import re
```

##**Importing the dataset**##

```python
file_path = 'Updated_diabetics.csv'
df = pd.read_csv(file_path)
df.head()
```

##**Standerdizing drug name combine with dosage**##

```python
# Standardize Drug Names by removing dosage information
def standardize_drugname(drug_name):
    return re.sub(r'\s*\d+\s*MG', '', drug_name, flags=re.IGNORECASE).strip()

df['Standardized Drug Name'] = df['Drug Name'].apply(standardize_drugname)

# Combine Standardized Drug Name with Dosage to create a unique identifier
df['Drug Identifier'] = df['Standardized Drug Name'] + ' ' + df['Dosage'].astype(str)

# Update Drug Name to ensure consistency
df['Drug Name'] = df['Drug Identifier']

#Drop tempory columns
df.drop(columns=['Standardized Drug Name', 'Drug Identifier'], inplace=True)
```

##**Output dataset**##

```python
from google.colab import files
df.to_csv('Standerdize_data_diabetics.csv', index=False)
files.download('Standerdize_data_diabetics.csv')
```
