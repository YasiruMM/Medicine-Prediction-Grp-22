```python
import pandas as pd
file_path= '/content/drive/My Drive/DSGP/Updated2_CardioDATA.csv'  # Replace with your actual file name
#Load the Excel file into a DataFrame
df = pd.read_csv(file_path)

# display head
df.head()
```

```python
# Sort by 'Drug Name', 'Date', and 'Sales' to keep the highest sales record
df_sorted = df.sort_values(by=['Drug Name', 'Date', 'Sales'], ascending=[True, True, False])

# Drop duplicates, keeping the first occurrence (which will be the one with the highest sales)
df_filtered = df_sorted.drop_duplicates(subset=['Drug Name', 'Date'], keep='first')

# Display the result after handling duplicates
print("\nDataFrame after filtering duplicates (keeping highest sales):")
print(df_filtered)
```

```python
from google.colab import files
# Save the processed data to a CSV file in Colab's local environment
output_csv_path = '/content/cardi_DUPLI_Handle.csv'
df_filtered.to_csv(output_csv_path, index=False)
# Download the CSV file to your local machine
files.download(output_csv_path)
```
