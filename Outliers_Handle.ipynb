```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path= '/content/drive/My Drive/DSGP/Name_Handle.csv'
#Load the Excel file into a DataFrame
df = pd.read_csv(file_path)
```

# Scatter plot to detect Purchase vs Retail Prices Outliers

```python
# Scatter Plot: Retail Price vs. Purchase Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Purchase Price'], y=df['Retail Price'])
plt.title("Scatter Plot: Retail Price vs. Purchase Price")
plt.xlabel("Purchase Price")
plt.ylabel("Retail Price")
plt.grid(True)
plt.show()
```

# Box plot for Sales Outliers

```python
# Box Plot: Sales
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Sales'])
plt.title("Box Plot: Sales")
plt.xlabel("Sales")
plt.show()
```

Winsorization for Outliers

```python
# create a function for capping
def cap_values(column):
    lower_cap = column.quantile(0.05)
    upper_cap = column.quantile(0.95)
    return column.clip(lower=lower_cap, upper=upper_cap)
```

```python
# Apply capping to Purchase Price, Retail Price, and Sales
df['Purchase Price'] = cap_values(df['Purchase Price'])
df['Retail Price'] = cap_values(df['Retail Price'])
df['Sales'] = cap_values(df['Sales'])
```

```python
# visualize outliers
def plot_outliers(df, feature, title):
    plt.figure(figsize=(10, 5))
    plt.boxplot(df[feature])
    plt.title(title)
    plt.show()
```

```python
# Plot Purchase Price, Retail Price, and Sales after capping
plot_outliers(df, 'Purchase Price', 'Purchase Price - After Capping')
plot_outliers(df, 'Retail Price', 'Retail Price - After Capping')
plot_outliers(df, 'Sales', 'Sales - After Capping')
```
